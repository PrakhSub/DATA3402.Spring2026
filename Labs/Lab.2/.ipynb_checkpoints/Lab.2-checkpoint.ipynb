*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
# Write your solution here
def make_board(n):
    board = []
    for _ in range(n):
        row = [0] * n
        board.append(row)
    return board
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
# Write your solution here
def draw_board_from_matrix(board):
    n = len(board)
    m = len(board[0])

    for i in range(n):
        print(" ---" * m)

        row_line = ""
        for j in range(m):
            value = board[i][j]
            if value == 0:
                symbol = " "
            elif value == 1:
                symbol = "X"
            else:
                symbol = "O"

            row_line = row_line + "|" + " " + symbol + " "

        row_line = row_line + "|"
        print(row_line)

    print(" ---" * m)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def check_winner(board):
    n = len(board)
    for i in range(n):
        first = board[i][0]
        if first != 0:
            same = True
            for j in range(1, n):
                if board[i][j] != first:
                    same = False
                    break
            if same:
                return first

    for j in range(n):
        first = board[0][j]
        if first != 0:
            same = True
            for i in range(1, n):
                if board[i][j] != first:
                    same = False
                    break
            if same:
                return first
                
    first = board[0][0]
    if first != 0:
        same = True
        for k in range(1, n):
            if board[k][k] != first:
                same = False
                break
        if same:
            return first
            
    first = board[0][n - 1]
    if first != 0:
        same = True
        for k in range(1, n):
            if board[k][n - 1 - k] != first:
                same = False
                break
        if same:
            return first
            
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                return -1

    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def print_board(board):
    for row in board:
        print(row)
    print()

def place_mark(board, player, row, col):
    if board[row][col] != 0:
        print("Position already filled!")
        return False
    else:
        board[row][col] = player
        return True
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
# Write your solution here
def draw_board_from_matrix(board):
    n = len(board)
    m = len(board[0])

    label_line = "   "
    for col in range(m):
        label_line = label_line + " " + str(col + 1) + "  "
    print(label_line)

    for i in range(n):
        row_label = chr(ord('A') + i)
        print("   " + " ---" * m)

        row_line = row_label + " "
        for j in range(m):
            value = board[i][j]
            if value == 0:
                symbol = " "
            elif value == 1:
                symbol = "X"
            else:
                symbol = "O"
            row_line = row_line + "|" + " " + symbol + " "
        row_line = row_line + "|"
        print(row_line)

    print("   " + " ---" * m)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here
def place_mark_from_string(board, player, location):

    letter = location[0]
    row = ord(letter) - ord('A')

    number = location[1]
    col = int(number) - 1

    success = place_mark(board, player, row, col)
    return success
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write your solution here
def play_tic_tac_toe(n):
    board = make_board(n)
    while True:
        print("Player 1 (X) turn:")
        player_turn(board, 1)

        winner = check_winner(board)
        if winner != -1:
            draw_board_from_matrix(board)
            if winner == 1:
                print("Player 1 (X) wins!")
            elif winner == 0:
                print("It's a draw!")
            break

        print("Player 2 (O) turn:")
        player_turn(board, 2)

        winner = check_winner(board)
        if winner != -1:
            draw_board_from_matrix(board)
            if winner == 2:
                print("Player 2 (O) wins!")
            elif winner == 0:
                print("It's a draw!")
            break
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
# Write your solution here
def computer_turn(board, computer_player):
    n = len(board)
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer_player
                if check_winner(board) == computer_player:
                    draw_board_from_matrix(board)
                    return True
                board[i][j] = 0

    human_player = 3 - computer_player
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = human_player
                if check_winner(board) == human_player:
                    board[i][j] = computer_player
                    draw_board_from_matrix(board)
                    return True
                board[i][j] = 0

    center_i, center_j = n//2, n//2
    if board[center_i][center_j] == 0:
        board[center_i][center_j] = computer_player
        draw_board_from_matrix(board)
        return True

    corners = [(0,0), (0,n-1), (n-1,0), (n-1,n-1)]
    for i, j in corners:
        if board[i][j] == 0:
            board[i][j] = computer_player
            draw_board_from_matrix(board)
            return True

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer_player
                draw_board_from_matrix(board)
                return True


def play_vs_computer(n):
    board = make_board(n)

    while True:
        print("\nYour turn (X):")
        player_turn(board, 1)

        winner = check_winner(board)
        if winner != -1:
            draw_board_from_matrix(board)
            if winner == 1:
                print("You win!")
            elif winner == 0:
                print("Draw!")
            break

        print("\nComputer thinking...")
        computer_turn(board, 2)

        winner = check_winner(board)
        if winner != -1:
            draw_board_from_matrix(board)
            if winner == 2:
                print("Computer wins!")
            elif winner == 0:
                print("Draw!")
            break
````

````python
# Test your solution here
play_vs_computer(3)
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def evaluate(board, maximizing_player):
    winner = check_winner(board)
    if winner == maximizing_player:
        return 10
    elif winner == 3 - maximizing_player:
        return -10
    elif winner == 0:
        return 0
    else:
        my_marks = sum(sum(1 for col in row if col == maximizing_player) for row in board)
        opp_marks = sum(sum(1 for col in row if col == 3 - maximizing_player) for row in board)
        return my_marks - opp_marks

def minimax(board, depth, max_depth, current_player, maximizing_player):
    if depth >= max_depth:
        return evaluate(board, maximizing_player)

    winner = check_winner(board)
    if winner != -1:
        return evaluate(board, maximizing_player)

    n = len(board)
    best = -999 if current_player == maximizing_player else 999

    for row in range(n):
        for col in range(n):
            if board[row][col] == 0:
                board[row][col] = current_player
                score = minimax(board, depth + 1, max_depth,
                               3 - current_player, maximizing_player)
                board[row][col] = 0

                if current_player == maximizing_player:
                    best = max(best, score)
                else:
                    best = min(best, score)

    return best

def computer_turn(board, maximizing_player=2, max_depth=6):
    n = len(board)
    best_score = -999
    best_move = None

    for row in range(n):
        for col in range(n):
            if board[row][col] == 0:
                board[row][col] = maximizing_player
                score = minimax(board, 0, max_depth,
                               3 - maximizing_player, maximizing_player)
                board[row][col] = 0

                if score > best_score:
                    best_score = score
                    best_move = (row, col)

    if best_move:
        row, col = best_move
        place_mark(board, maximizing_player, row, col)
        return True
    return False


def play_vs_computer(n=3, computer_player=2, max_depth=9):
    board = make_board(n)
    current_player = 1

    while True:
        draw_board_from_matrix(board)
        winner = check_winner(board)

        if winner != -1:
            if winner == 1:
                print("You win!")
            elif winner == 2:
                print("Computer wins!")
            else:
                print("Draw!")
            break

        if current_player == 1:
            player_turn(board, 1)
        else:
            print("Computer thinking...")
            computer_turn(board, computer_player, max_depth)

        current_player = 3 - current_player
````

````python
# Test your solution here
play_vs_computer(3, max_depth=9)
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here
def make_board(n):
    return [[0 for _ in range(n)] for _ in range(n)]

def check_winner(board):
    n = len(board)
    for row in board:
        if all(cell == 1 for cell in row): return 1
        if all(cell == 2 for cell in row): return 2
    for c in range(n):
        col = [board[r][c] for r in range(n)]
        if all(cell == 1 for cell in col): return 1
        if all(cell == 2 for cell in col): return 2
    diag1 = [board[i][i] for i in range(n)]
    if all(cell == 1 for cell in diag1): return 1
    if all(cell == 2 for cell in diag1): return 2
    diag2 = [board[i][n-1-i] for i in range(n)]
    if all(cell == 1 for cell in diag2): return 1
    if all(cell == 2 for cell in diag2): return 2
    return 0

def is_full(board):
    return all(cell != 0 for row in board for cell in row)

def get_available(board):
    n = len(board)
    return [(i,j) for i in range(n) for j in range(n) if board[i][j] == 0]

def minimax(board, player, depth):
    winner = check_winner(board)
    if winner == 1: return 1
    if winner == 2: return -1
    if is_full(board) or depth == 0: return 0
    
    best = -2 if player == 1 else 2
    for i,j in get_available(board):
        board[i][j] = player
        score = minimax(board, 3-player, depth-1)
        board[i][j] = 0
        best = max(score, best) if player == 1 else min(score, best)
    return best

def best_move(board, player, max_depth):
    best_score = -2 if player == 1 else 2
    best_pos = None
    
    for i,j in get_available(board):
        board[i][j] = player
        score = minimax(board, 3-player, max_depth-1)
        board[i][j] = 0
        
        if (player == 1 and score > best_score) or (player == 2 and score < best_score):
            best_score = score
            best_pos = (i,j)
    
    return best_pos

def play_game(n, depth_smart=3, depth_dumb=1):
    board = make_board(n)
    current = 1
    while True:
        depth = depth_smart if current == 1 else depth_dumb
        move = best_move(board, current, depth)
        i,j = move
        board[i][j] = current
        
        winner = check_winner(board)
        if winner: return winner
        if is_full(board): return 0
        
        current = 3 - current

def run_tournament(grid_sizes=[3,4,5], num_games=10):
    print("🎮 AI Tic-Tac-Toe Tournament (Smart depth=3 vs Dumb depth=1)")
    print("="*60)
    for n in grid_sizes:
        smart_wins = sum(1 for _ in range(num_games) if play_game(n) == 1)
        win_rate = (smart_wins/num_games)*100
        print(f"📊 {n}x{n} grid: {smart_wins}/{num_games} wins = {win_rate:.0f}% win rate")
````

````python
# Test your solution here
run_tournament()
````
