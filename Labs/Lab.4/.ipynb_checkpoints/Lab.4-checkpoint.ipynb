1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.count = 0

    def increment(self):
        if self.count >= self.max_value:
            print("Error: Counter has reached its maximum value!")
        else:
            self.count += 1

    def reset(self):
        self.count = 0

my_counter = Counter(3)
my_counter.increment()
print("Count after 1 increment:", my_counter.count)

my_counter.increment()
my_counter.increment()
print("Count after 3 increments:", my_counter.count)

print("Trying to increment again...")
my_counter.increment() 

my_counter.reset()
print("Count after reset:", my_counter.count)
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, max_value):
        self.__max_value = max_value
        self.__count = 0

    def increment(self):
        if self.is_at_max():
            print("Error: Counter has reached its maximum value!")
        else:
            self.__count += 1

    def reset(self):
        self.__count = 0
    
    def get_value(self):
        return self.__count

    def get_max_value(self):
        return self.__max_value

    def is_at_max(self):
        return self.__count >= self.__max_value
    
my_counter = Counter(3)

print("Initial value:", my_counter.get_value())
print("Max allowed value:", my_counter.get_max_value())
print("Is it at max right now?", my_counter.is_at_max())

print("\nIncrementing 3 times...")
my_counter.increment()
my_counter.increment()
my_counter.increment()

print("Current value:", my_counter.get_value())
print("Is it at max right now?", my_counter.is_at_max())

print("\nTrying to increment one more time...")
my_counter.increment()

print("\nResetting the counter...")
my_counter.reset()
print("Value after reset:", my_counter.get_value())
```

```python
class Rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    
    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y
    
    def get_area(self):
        return self.__length * self.__width

    def get_perimeter(self):
        return 2 * (self.__length + self.__width)

my_rect = Rectangle(length=5, width=3, x=10, y=20)

print("--- Rectangle Details ---")
print("Length:", my_rect.get_length())
print("Width:", my_rect.get_width())
print("X Coordinate:", my_rect.get_x())
print("Y Coordinate:", my_rect.get_y())

print("\n--- Computations ---")
print("Area (5 * 3):", my_rect.get_area())
print("Perimeter (2 * (5 + 3)):", my_rect.get_perimeter())

print(my_rect.__length)
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle:
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
    
    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y
    
    def get_area(self):
        return math.pi * (self.__radius ** 2)

    def get_perimeter(self):
        return 2 * math.pi * self.__radius

my_circle = Circle(radius=5, x=0, y=0)

print("--- Circle Details ---")
print("Radius:", my_circle.get_radius())
print("Center X Coordinate:", my_circle.get_x())
print("Center Y Coordinate:", my_circle.get_y())
print("\n--- Computations ---")
print("Area (pi * r^2):", my_circle.get_area())
print("Perimeter/Circumference (2 * pi * r):", my_circle.get_perimeter())

print(my_circle.__radius)
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math

class Shape:
    def get_x(self):
        raise NotImplementedError("Subclass must implement get_x()")

    def get_y(self):
        raise NotImplementedError("Subclass must implement get_y()")

    def get_area(self):
        raise NotImplementedError("Subclass must implement get_area()")

    def get_perimeter(self):
        raise NotImplementedError("Subclass must implement get_perimeter()")

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_perimeter(self):
        return self.__side_a + self.__side_b + self.__side_c

    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_side_a(self):
        return self.__side_a

    def get_side_b(self):
        return self.__side_b

    def get_side_c(self):
        return self.__side_c

print("--- Testing Triangle ---")
my_triangle = Triangle(side_a=3, side_b=4, side_c=5, x=2, y=5)

print("X, Y Coordinates:", my_triangle.get_x(), ",", my_triangle.get_y())
print("Side A:", my_triangle.get_side_a())
print("Side B:", my_triangle.get_side_b())
print("Side C:", my_triangle.get_side_c())

print("\n--- Computations ---")
print("Perimeter (3 + 4 + 5):", my_triangle.get_perimeter(), "(Expected: 12)")
print("Area (via Heron's formula):", my_triangle.get_area(), "(Expected: 6.0)")

print("\n--- Testing the Virtual Base Class ---")
generic_shape = Shape()
try:
    generic_shape.get_area()
except NotImplementedError as e:
    print("Success! The base class prevented us from running a virtual method:")
    print("Error Message ->", e)
```

```python
import math

class Shape:
    def get_x(self):
        raise NotImplementedError()

    def get_y(self):
        raise NotImplementedError()

    def get_area(self):
        raise NotImplementedError()

    def get_perimeter(self):
        raise NotImplementedError()

    def get_perimeter_points(self):
        raise NotImplementedError("Subclass must implement get_perimeter_points()")

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        
        for i in range(4):
            points.append((self.__x + i * (self.__width / 4), self.__y))
            
        for i in range(4):
            points.append((self.__x + self.__width, self.__y + i * (self.__length / 4)))
            
        for i in range(4):
            points.append((self.__x + self.__width - i * (self.__width / 4), self.__y + self.__length))
            
        for i in range(4):
            points.append((self.__x, self.__y + self.__length - i * (self.__length / 4)))
            
        return points

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        for i in range(16):
            angle = i * (2 * math.pi / 16)
            px = self.__x + self.__radius * math.cos(angle)
            py = self.__y + self.__radius * math.sin(angle)
            points.append((round(px, 4), round(py, 4)))
        return points

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_side_a(self): return self.__side_a
    def get_side_b(self): return self.__side_b
    def get_side_c(self): return self.__side_c
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c

    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_perimeter_points(self):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)

        points = []
        for i in range(6):
            t = i / 6
            points.append((round(v1_x + t*(v2_x-v1_x), 4), round(v1_y + t*(v2_y-v1_y), 4)))
            
        for i in range(5):
            t = i / 5
            points.append((round(v2_x + t*(v3_x-v2_x), 4), round(v2_y + t*(v3_y-v2_y), 4)))
            
        for i in range(5):
            t = i / 5
            points.append((round(v3_x + t*(v1_x-v3_x), 4), round(v3_y + t*(v1_y-v3_y), 4)))
            
        return points

print("--- Testing Rectangle Points ---")
my_rect = Rectangle(length=4, width=4, x=0, y=0)
rect_points = my_rect.get_perimeter_points()
print(f"Rectangle generated {len(rect_points)} points.")
print("First 3 points:", rect_points[:3])

print("\n--- Testing Circle Points ---")
my_circle = Circle(radius=5, x=0, y=0)
circle_points = my_circle.get_perimeter_points()
print(f"Circle generated {len(circle_points)} points.")
print("First 3 points:", circle_points[:3])

print("\n--- Testing Triangle Points ---")
my_triangle = Triangle(side_a=3, side_b=4, side_c=5, x=0, y=0)
tri_points = my_triangle.get_perimeter_points()
print(f"Triangle generated {len(tri_points)} points.")
print("First 3 points:", tri_points[:3])

print("\n--- Testing Base Class ---")
generic_shape = Shape()
try:
    generic_shape.get_perimeter_points()
except NotImplementedError as e:
    print("Success! Virtual method protected:", e)
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
import math

class Shape:
    def get_x(self): raise NotImplementedError()
    def get_y(self): raise NotImplementedError()
    def get_area(self): raise NotImplementedError()
    def get_perimeter(self): raise NotImplementedError()
    def get_perimeter_points(self): raise NotImplementedError()
    def is_inside(self, target_x, target_y): raise NotImplementedError()

    def overlaps(self, other):
        for pt in self.get_perimeter_points():
            if other.is_inside(pt[0], pt[1]):
                return True
                
        for pt in other.get_perimeter_points():
            if self.is_inside(pt[0], pt[1]):
                return True                
        return False

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        for i in range(4): points.append((self.__x + i * (self.__width / 4), self.__y))
        for i in range(4): points.append((self.__x + self.__width, self.__y + i * (self.__length / 4)))
        for i in range(4): points.append((self.__x + self.__width - i * (self.__width / 4), self.__y + self.__length))
        for i in range(4): points.append((self.__x, self.__y + self.__length - i * (self.__length / 4)))
        return points

    def is_inside(self, target_x, target_y):
        in_x = self.__x <= target_x <= (self.__x + self.__width)
        in_y = self.__y <= target_y <= (self.__y + self.__length)
        return in_x and in_y

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        for i in range(16):
            angle = i * (2 * math.pi / 16)
            px = self.__x + self.__radius * math.cos(angle)
            py = self.__y + self.__radius * math.sin(angle)
            points.append((px, py))
        return points

    def is_inside(self, target_x, target_y):
        distance_squared = (target_x - self.__x)**2 + (target_y - self.__y)**2
        return distance_squared <= (self.__radius ** 2)

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c
    
    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_perimeter_points(self):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)

        points = []
        for i in range(6):
            t = i / 6
            points.append((v1_x + t*(v2_x-v1_x), v1_y + t*(v2_y-v1_y)))
        for i in range(5):
            t = i / 5
            points.append((v2_x + t*(v3_x-v2_x), v2_y + t*(v3_y-v2_y)))
        for i in range(5):
            t = i / 5
            points.append((v3_x + t*(v1_x-v3_x), v3_y + t*(v1_y-v3_y)))
        return points

    def __calc_area_from_points(self, x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2.0)

    def is_inside(self, target_x, target_y):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)

        total_area = self.get_area()
        area1 = self.__calc_area_from_points(target_x, target_y, v2_x, v2_y, v3_x, v3_y)
        area2 = self.__calc_area_from_points(v1_x, v1_y, target_x, target_y, v3_x, v3_y)
        area3 = self.__calc_area_from_points(v1_x, v1_y, v2_x, v2_y, target_x, target_y)

        return abs(total_area - (area1 + area2 + area3)) < 1e-5

print("--- Testing Overlaps ---")

rect1 = Rectangle(length=4, width=4, x=0, y=0)

rect2 = Rectangle(length=4, width=4, x=2, y=2)

rect3 = Rectangle(length=2, width=2, x=10, y=10)

circ = Circle(radius=5, x=0, y=0)

print("rect1 overlaps rect2?", rect1.overlaps(rect2), "(Expected: True)")
print("rect1 overlaps rect3?", rect1.overlaps(rect3), "(Expected: False)")
print("circ overlaps rect1?", circ.overlaps(rect1), "(Expected: True)")
print("circ overlaps rect3?", circ.overlaps(rect3), "(Expected: False)")
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def get_x(self): raise NotImplementedError()
    def get_y(self): raise NotImplementedError()
    def get_area(self): raise NotImplementedError()
    def get_perimeter(self): raise NotImplementedError()
    def get_perimeter_points(self): raise NotImplementedError()
    def is_inside(self, target_x, target_y): raise NotImplementedError()
    def overlaps(self, other):
        for pt in self.get_perimeter_points():
            if other.is_inside(pt[0], pt[1]): return True
        for pt in other.get_perimeter_points():
            if self.is_inside(pt[0], pt[1]): return True
        return False
    def paint(self, canvas, char='*'): raise NotImplementedError()

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        for i in range(4): points.append((self.__x + i * (self.__width / 4), self.__y))
        for i in range(4): points.append((self.__x + self.__width, self.__y + i * (self.__length / 4)))
        for i in range(4): points.append((self.__x + self.__width - i * (self.__width / 4), self.__y + self.__length))
        for i in range(4): points.append((self.__x, self.__y + self.__length - i * (self.__length / 4)))
        return points

    def is_inside(self, target_x, target_y):
        in_x = self.__x <= target_x <= (self.__x + self.__width)
        in_y = self.__y <= target_y <= (self.__y + self.__length)
        return in_x and in_y
        
    def paint(self, canvas, char='*'):
        x = int(self.__x)
        y = int(self.__y)
        w = int(self.__width)
        h = int(self.__length)
        for i in range(w + 1):
            if 0 <= y < canvas.height and 0 <= x + i < canvas.width:
                canvas.set_pixel(y, x + i, char=char)
            if 0 <= y + h < canvas.height and 0 <= x + i < canvas.width:
                canvas.set_pixel(y + h, x + i, char=char)
        for i in range(h + 1):
            if 0 <= y + i < canvas.height and 0 <= x < canvas.width:
                canvas.set_pixel(y + i, x, char=char)
            if 0 <= y + i < canvas.height and 0 <= x + w < canvas.width:
                canvas.set_pixel(y + i, x + w, char=char)

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        for i in range(16):
            angle = i * (2 * math.pi / 16)
            px = self.__x + self.__radius * math.cos(angle)
            py = self.__y + self.__radius * math.sin(angle)
            points.append((px, py))
        return points

    def is_inside(self, target_x, target_y):
        distance_squared = (target_x - self.__x)**2 + (target_y - self.__y)**2
        return distance_squared <= (self.__radius ** 2)

    def paint(self, canvas, char='*'):
        r = self.__radius
        cx = self.__x
        cy = self.__y
        circumference = 2 * math.pi * r
        points_to_draw = int(circumference * 3) 
        for i in range(points_to_draw):
            angle = i * (2 * math.pi / points_to_draw)
            px = int(round(cx + r * math.cos(angle)))
            py = int(round(cy + r * math.sin(angle)))
            if 0 <= py < canvas.height and 0 <= px < canvas.width:
                canvas.set_pixel(py, px, char=char)

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c
    
    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_perimeter_points(self):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)

        points = []
        for i in range(6):
            t = i / 6
            points.append((v1_x + t*(v2_x-v1_x), v1_y + t*(v2_y-v1_y)))
        for i in range(5):
            t = i / 5
            points.append((v2_x + t*(v3_x-v2_x), v2_y + t*(v3_y-v2_y)))
        for i in range(5):
            t = i / 5
            points.append((v3_x + t*(v1_x-v3_x), v3_y + t*(v1_y-v3_y)))
        return points

    def __calc_area_from_points(self, x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2.0)

    def is_inside(self, target_x, target_y):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)

        total_area = self.get_area()
        area1 = self.__calc_area_from_points(target_x, target_y, v2_x, v2_y, v3_x, v3_y)
        area2 = self.__calc_area_from_points(v1_x, v1_y, target_x, target_y, v3_x, v3_y)
        area3 = self.__calc_area_from_points(v1_x, v1_y, v2_x, v2_y, target_x, target_y)

        return abs(total_area - (area1 + area2 + area3)) < 1e-5

    def paint(self, canvas, char='*'):
        v1_x, v1_y = self.__x, self.__y
        v2_x, v2_y = self.__x + self.__side_a, self.__y
        angle_c = math.acos((self.__side_a**2 + self.__side_b**2 - self.__side_c**2) / (2 * self.__side_a * self.__side_b))
        v3_x = self.__x + self.__side_b * math.cos(angle_c)
        v3_y = self.__y + self.__side_b * math.sin(angle_c)
        
        def draw_line(x_start, y_start, x_end, y_end):
            dist = max(abs(x_end - x_start), abs(y_end - y_start))
            dist = int(math.ceil(dist))
            for i in range(dist + 1):
                t = i / dist if dist > 0 else 0
                px = int(round(x_start + t * (x_end - x_start)))
                py = int(round(y_start + t * (y_end - y_start)))
                if 0 <= py < canvas.height and 0 <= px < canvas.width:
                    canvas.set_pixel(py, px, char=char)

        draw_line(v1_x, v1_y, v2_x, v2_y)
        draw_line(v2_x, v2_y, v3_x, v3_y)
        draw_line(v3_x, v3_y, v1_x, v1_y)

class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def paint(self, canvas, char='*'):
        for s in self.shapes:
            s.paint(canvas, char=char)


my_canvas = Canvas(40, 20)

my_rect = Rectangle(length=6, width=12, x=2, y=2)
my_circle = Circle(radius=4, x=28, y=8)
my_triangle = Triangle(side_a=8, side_b=10, side_c=12, x=10, y=10)

my_compound = CompoundShape([my_rect, my_circle, my_triangle])
my_compound.paint(my_canvas, char='@')

my_canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = dict()
        self.shape_names = list()
        
    def add_shape(self, shape):
        if not hasattr(shape, 'name') or shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name] = shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self, canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base = "shape"
        name = name_base + "_0"
        
        i = 1
        while name in self.shapes:
            name = name_base + "_" + str(i)
            i += 1
            
        return name

def rect_set_width(self, new_width):
    self._Rectangle__width = new_width
    
def rect_set_length(self, new_length):
    self._Rectangle__length = new_length

Rectangle.set_width = rect_set_width
Rectangle.set_length = rect_set_length

print("--- Initial Drawing ---")
my_canvas = Canvas(40, 20)
rd = RasterDrawing()

rect1 = Rectangle(length=4, width=8, x=2, y=2)
circ1 = Circle(radius=3, x=25, y=5)

rd.add_shape(rect1)
rd.add_shape(circ1)

rd.paint(my_canvas)
my_canvas.display()

print("\n--- Modified Drawing ---")
rd.shapes["shape_0"].set_width(18)
rd.shapes["shape_0"].set_length(8)

tri1 = Triangle(side_a=6, side_b=6, side_c=8, x=2, y=12)
rd.add_shape(tri1)

rd.update(my_canvas)
my_canvas.display()
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
def rect_repr(self):
    return f"Rectangle({self.get_length()}, {self.get_width()}, {self.get_x()}, {self.get_y()})"

Rectangle.__repr__ = rect_repr

def circle_repr(self):
    return f"Circle({self.get_radius()}, {self.get_x()}, {self.get_y()})"

Circle.__repr__ = circle_repr

def triangle_repr(self):
    return f"Triangle({self.get_side_a()}, {self.get_side_b()}, {self.get_side_c()}, {self.get_x()}, {self.get_y()})"

Triangle.__repr__ = triangle_repr

def compound_repr(self):
    return f"CompoundShape({repr(self.shapes)})"

CompoundShape.__repr__ = compound_repr


class RasterDrawing:
    def __init__(self, initial_shapes=None):
        self.shapes = dict()
        self.shape_names = list()
        if initial_shapes:
            for s in initial_shapes:
                self.add_shape(s)
        
    def add_shape(self, shape):
        if not hasattr(shape, 'name') or shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name] = shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self, canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base = "shape"
        name = name_base + "_0"
        
        i = 1
        while name in self.shapes:
            name = name_base + "_" + str(i)
            i += 1
            
        return name

    def __repr__(self):
        shapes_list = [repr(s) for s in self.shapes.values()]
        return f"RasterDrawing([{', '.join(shapes_list)}])"

    def save(self, filename):
        f = open(filename, "w")
        f.write(repr(self))
        f.close()


def load_drawing(filename):
    f = open(filename, "r")
    tmp = eval(f.read())
    f.close()
    return tmp


my_canvas = Canvas(40, 20)
rd = RasterDrawing()

rect1 = Rectangle(4, 8, 2, 2)
circ1 = Circle(3, 25, 5)

rd.add_shape(rect1)
rd.add_shape(circ1)

rd.save("test_drawing.txt")

loaded_rd = load_drawing("test_drawing.txt")

loaded_rd.update(my_canvas)
my_canvas.display()
```
